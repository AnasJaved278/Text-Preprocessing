```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# create output folder
os.makedirs("outputs", exist_ok=True)

print("Libraries imported successfully!")
```

```python
df = pd.read_csv("amazon.csv")

print("Dataset loaded successfully!")
print("Shape of dataset:", df.shape)
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
# Fill missing numerical values with mean
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill missing text values with mode
text_cols = df.select_dtypes(include="object").columns
for col in text_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("Missing values handled!")
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z ]", "", text)
    return text

df["product_name"] = df["product_name"].apply(clean_text)

print("Text data cleaned!")
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for col in text_cols:
    df[col] = encoder.fit_transform(df[col])

print("Categorical data encoded!")
```

```python
df.dtypes
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(numeric_cols)
```

```python
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("Numerical data scaled between 0 and 1!")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df[numeric_cols[0]])
plt.title("Distribution of Sales")
plt.savefig("outputs/sales_distribution.png")
plt.show()
```

```python
df.to_csv("outputs/cleaned_amazon_sales.csv", index=False)
print("Cleaned dataset saved!")
```
